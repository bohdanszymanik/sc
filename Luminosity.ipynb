```r
# get sensors
library(RCurl)
tJson <- postForm("https://api.necsmart.city/v0.1/wellington-token")
library(jsonlite)
token <- fromJSON(tJson)$token

sensors <- fromJSON(paste0("https://api.necsmart.city/v0.1/sensors?token=", token))
```

```r
# let's use a luminosity measure as an example
alux <- fromJSON(paste0("https://api.necsmart.city/v0.1/sensors/184032/events?token=", token))
require(parsedate)
# a little simpler dataframe
lux <- data.frame(datetime = parse_date(alux$events$reading_at), value = as.numeric(alux$events$reading_value))

require(ggplot2)
ggplot(lux, aes(x=datetime, y=value)) +
  geom_point(alpha=0.5, size=1) +
  geom_line()
```
